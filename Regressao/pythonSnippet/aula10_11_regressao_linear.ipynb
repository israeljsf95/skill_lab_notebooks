```python
X = df
y = target['MEDV']
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 50))

for i, col in enumerate(X.columns):
    plt.subplot(7, 2, i+1)
    x = X[col]
    y = y
    plt.plot(x, y, 'o')
    
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('Pre√ßos')
    
```

```python
X = df['RM'].values.reshape(-1, 1)
y = target['MEDV']

lm = linear_model.LinearRegression()
modelo_treinado = lm.fit(X, y)
```
